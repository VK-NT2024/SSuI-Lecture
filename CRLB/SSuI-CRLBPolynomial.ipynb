# Cramer-Rao Lower Bound

## Allgemeine Herleitung

Die Cramer-Rao Lower Bound (CRLB) ist die unter Schranke für die Schätzvarianz erwartungstreuer Schätzer. Sie basiert auf der Likelihood-Funktion $p({\mathbf y};{\mathbf \theta})$, wobei $\mathbf y$ der Messvektor bestehend aus $N$ Messwerten $y(t_n)$ und $\mathbf \theta = \begin{bmatrix} \theta_0 & \cdots & \theta_{P-1} \end{bmatrix}^{\mathrm{T}}$ der zu schätzende Parametervektor sind. Die Fisher-Informationsmatrix ${\mathbf J}({\mathbf \theta})$ ergibt sich aus dem negativen Erwartungswert der Hesse-Matrix der Log-Likelihood-Funktion.

\begin{equation}
    {\mathbf J}({\mathbf \theta})
    = - {\mathrm E} \left\{ \begin{bmatrix}  
        \frac{\partial^2 \log p({\mathbf y};{\mathbf \theta})}{\partial \theta_0^2}  & \cdots &
        \frac{\partial^2 \log p({\mathbf y};{\mathbf \theta})}{\partial \theta_0 \partial \theta_{P-1}} \\
       \vdots & \ddots & \vdots \\
       \frac{\partial^2 \log p({\mathbf y};{\mathbf \theta})}{\partial \theta_{P-1} \partial \theta_0} 
       & \cdots & \frac{\partial^2 \log p({\mathbf y};{\mathbf \theta})}{\partial \theta_{P-1}^2} 
    \end{bmatrix} \right\} .
\end{equation}

Berechnet man die Inverse ${\mathbf J}({\mathbf \theta})^{-1}$ der Fisher-Informationsmatrix, erhält man auf der Diagonalen die unteren Schranken der Schätzvarianzen für die einzelnen Parameter. Es gilt

\begin{equation}
    \mathrm{var}(\hat{\theta}_p) = \left[ {\mathbf J}^{-1}({\mathbf \theta}) \right]_{p,p} \; .
\end{equation}

Die Gesamtvarianz für den Parametervektor $\mathbf \theta$ ergibt sich aus der Summe der einzelnen Varianzen.
Ferner unterscheiden wir analoge und digitale Signale sowie Mischformen.

\begin{equation}
    \mathrm{var}(\hat{\mathbf \theta}) 
    = \sum_{p=0}^{P-1} \mathrm{var}(\hat{\theta}_p) 
    = \left\| \mathrm{diag} \left( {\mathbf J}^{-1}({\mathbf \theta}) \right) \right\|_1 \; .
\end{equation}


## Spezieller Fall für Polynome

Die Messwerte $y(t_n)$ setzen sich häufig aus dem Informationssignal $x(t_n)$ und additivem Rauschen $w(t_n)$ zusammen, d.h. es gilt $y(t_n)=x(t_n)+w(t_n)$. Im nun betrachteten Fall soll das Informationssignal die Form

\begin{equation}
     x(t_n) = \sum_{p=0}^{P-1} \theta_p t_n^p
\end{equation}

besitzen, es ist also durch $P$ Parameter $\theta_p$ und die Messzeitpunkte $t_n$ mit $0 \leq n < N$ definiert. Vektoriell lässt sich der Zusammenhang durch $\mathbf{x} = \mathbf{H} \cdot \mathbf{\theta}$ beschreiben, wobei

\begin{align}
    \mathbf{H}
    &= \begin{bmatrix}
         1 & t_0 & \cdots & t_0^{P-1} \\
         \vdots & \vdots &  & \vdots \\
         1 & t_{N-1} & \cdots & t_{N-1}^{P-1}
       \end{bmatrix}
\end{align}

eine Vandermonde-Matrix ist. Für weißes gaußverteiltes Rauschen mit der mittleren Leistung $\sigma_{\cal W}^2$ lautet die Likelihood-Funktion

\begin{align}
    p(\mathbf{y};\mathbf{\theta}) 
    &= \frac{1}{\sqrt{\mathrm{det}(2 \pi \Phi_{\cal WW})}} \cdot 
        e^{- \frac{1}{2} (\mathbf{y} - \mathbf{H \theta})^{\mathrm{T}} \Phi_{\cal WW}^{-1} (\mathbf{y} - \mathbf{H \theta})} \\
    &= (2 \pi \sigma_{\cal W}^2)^{-N/2} \cdot 
        e^{- \frac{\| \mathbf{y} - \mathbf{H \theta} \|^2}{2 \sigma_{\cal W}^2} } \; .
\end{align}

Damit erhalten wie die Log-Likelihood-Funktion

\begin{align}
    \log p(\mathbf{y};\mathbf{\theta}) 
    &= \mathrm{const.} - \frac{1}{2 \sigma_{\cal W}^2} \cdot
        \| \mathbf{y} - \mathbf{H \theta} \|^2 \\
    &= \mathrm{const.} - \frac{1}{2 \sigma_{\cal W}^2} \cdot
        \sum_{n=0}^{N-1} \left| y(t_n) - \sum_{p=0}^{P-1} \theta_p t_n^p \right|^2 \; .
\end{align}

Die partielle Ableitung der Log-Likelihood-Funktion nach dem Parameter $\theta_i$ liefert 

\begin{align}
    \frac{\partial \log p(\mathbf{y};\mathbf{\theta})}{\partial \theta_i}
    &= - \frac{1}{2 \sigma_{\cal W}^2} \cdot \sum_{n=0}^{N-1} 
        \frac{\partial}{\partial \theta_i} 
        \left( |y(t_n)|^2 - 2 y(t_n) \sum_{p=0}^{P-1} \theta_p t_n^p 
        + \sum_{p=0}^{P-1} \sum_{p'=0}^{P-1} \theta_p t_n^p \theta_{p'} t_n^{p'} \right) \\
    &= \frac{1}{2 \sigma_{\cal W}^2} \cdot \sum_{n=0}^{N-1} 
        \left( 2 y(t_n) t_n^i - 2 \sum_{p=0}^{P-1} \theta_p t_n^{p+i} \right) \\
\end{align}

Die zweite partielle Ableitung nach $\theta_j$ liefert das Ergebnis
\begin{align}
    \frac{\partial^2 \log p(\mathbf{y};\mathbf{\theta})}{\partial \theta_i \partial \theta_j}
    &= \frac{1}{\sigma_{\cal W}^2} \cdot \sum_{n=0}^{N-1} 
       \frac{\partial}{\partial \theta_j}
       \left( y(t_n) t_n^i - \sum_{p=0}^{P-1} \theta_p t_n^{p+i} \right) \\
    &= - \frac{1}{\sigma_{\cal W}^2} \cdot \sum_{n=0}^{N-1} t_n^{i+j} \\
\end{align}

Damit lautet die Fisher-Informationsmatrix

\begin{equation}
    {\mathbf J}({\mathbf \theta})
    = \frac{1}{\sigma_{\cal W}^2} \cdot
      \begin{bmatrix}  
        \sum_n t_n^0 & \cdots & \sum_n t_n^{P-1} \\
        \vdots & \ddots & \vdots \\
        \sum_n t_n^{P-1} & \cdots & \sum_n t_n^{2(P-1)} 
    \end{bmatrix} .
\end{equation}

Für eine äquidistante Abtastung mit $t_n = n T_a$ ergibt sich die Fisher-Informationsmatrix zu

\begin{equation}
    {\mathbf J}({\mathbf \theta})
    = \frac{1}{\sigma_{\cal W}^2} \cdot
      \begin{bmatrix}  
        N & \cdots & \sum_n (nT_a)^{P-1} \\
        \vdots & \ddots & \vdots \\
        \sum_n (nT_a)^{P-1} & \cdots & \sum_n (nT_a)^{2(P-1)} 
    \end{bmatrix} .
\end{equation}

Für den konkreten Fall $P=2$, also insgesamt 3 Parameter, können wir folgende endlichen Reihen ausnutzen.

* $\sum_{n=0}^{N-1} n = \frac{(N-1)N}{2}$
* $\sum_{n=0}^{N-1} n^2 = \frac{(N-1)N(2N-1)}{6}$
* $\sum_{n=0}^{N-1} n^3 = \frac{(N-1)^2 N^2}{4}$
* $\sum_{n=0}^{N-1} n^4 = \frac{(N-1)N(2N-1)(3(N-1)^2+3(N-1)-1}{30} = \frac{(N-1)N(2N-1)(3N^2-3N-1}{30}$

Hiermit können die Summen in der Fisher-Informationsmatrix durch Funktionen von $N$ ersetzt werden und wir erhalten

\begin{equation}
    {\mathbf J}({\mathbf \theta})
    = \frac{1}{\sigma_{\cal W}^2} \cdot
      \begin{bmatrix}  
        N & T_a \frac{(N-1)N}{2} & T_a^2 \frac{(N-1)N(2N-1)}{6} \\
        T_a \frac{(N-1)N}{2} & T_a^2 \frac{(N-1)N(2N-1)}{6} & T_a^3  \frac{(N-1)^2 N^2}{4} \\
        T_a^2 \frac{(N-1)N(2N-1)}{6} & T_a^3 \frac{(N-1)^2 N^2}{4} & T_a^4 \frac{(N-1)N(2N-1)(3N^2-3N-1)}{30} 
    \end{bmatrix} .
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt

# number of parameters to be estimated (rank of polynomial + 1)
P = 3

# number of measurements
N = np.arange(5,101,5)

# sampling interval
Ta = 1e-1

# variance of noise
sigma2_N = 1.0

# Fisher information matrix
crlb = np.zeros((len(N),P))
crlbTotal = np.zeros((len(N),1))

for cntr, n in enumerate(N):
    term1 = n*(n-1) / 2 * Ta
    term2 = (n-1)*n*(2*n-1)/6 * Ta**2
    term3 = n**2*(n-1)**2 / 4 * Ta**3
    term4 = (n-1)*n*(2*n-1)*(3*n**2-3*n-1) / 30 * Ta**4

    J = np.array([[n, term1, term2],[term1, term2, term3],[term2, term3, term4]])
    J /= sigma2_N
    Jinv = np.linalg.inv(J)
    
    crlb[cntr,:] = np.diag(Jinv)
    crlbTotal[cntr] = np.sum(np.diag(Jinv))


fig_crlb = plt.figure(figsize=(9,5))

ax1 = fig_crlb.add_subplot(111)
ax1.semilogy(N, crlb[:,0], 'b:+',label='$p_0$')
ax1.semilogy(N, crlb[:,1], 'g:x',label='$p_1$')
ax1.semilogy(N, crlb[:,2], 'r:o',label='$p_2$')
ax1.semilogy(N, crlbTotal, 'k:s',label='$total$')
plt.xlabel('$N$')
plt.ylabel('CRLB')
plt.title('CRLB für Polynom mit $P=$' + str(P) + ' und AWGN')
plt.grid(True)
plt.legend()
plt.show()
```

Das obige Bild zeigt die minimal möglichen Varianzen der Parameter $\theta_p$ für erwartungstreue Schätzer sowie die Gesamtvarianz für den Vektor $\theta$. Der Parameter $\theta_0$ kann für kurze Messreihen am genauesten geschätzt werden, die Schätzvarianzen für die übrigen beiden Parameter sinken aber mit zunehmender Sequenzlänge schneller ab und sind ab $N=22$ bzw. $N=45$ genauer bestimmbar. Für lange Sequenzen dominiert also die Varianz $\mathrm{var}(\theta_0)$ die Gesamtvarianz $\mathrm{var}(\theta) = \sum_{p=0}^{P-1} \mathrm{var}(\theta_p)$.

## Schätzung der Sequenzen $x[n]$

Soll statt des Parametervektors $\mathbf{\theta}$ das Informationssignal $x[n]$ geschätzt werden, welches eine Funktion von $\mathbf{\theta}$ ist, muss die CRLB um den Jacobi-Vektor erweitert werden. Mit

\begin{equation}
  x(t_n) = f(\theta) = \sum_{p=0}^{P-1} \theta_p \cdot t_n^p
\end{equation}

gilt

\begin{equation}
  \frac{\partial f(\theta)}{\partial \theta_i} 
  = \begin{cases}
      1 & \text{für } i=0 \\
      t & \text{für } i=1 \\
      t^2 & \text{für } i=2 \\
    \end{cases}
\end{equation}

und wir die erweiterte Cramer-Rao-Lower-Bound

\begin{equation}
    \mathrm{var}(\hat{x}[n])
    = \begin{bmatrix} 1 \\ t_n \\ t_n^2 \end{bmatrix}^{\mathrm{T}}
        \cdot \mathbf{J}^{-1}(\mathbf{\theta}) \cdot
        \begin{bmatrix} 1 \\ t_n \\ t_n^2 \end{bmatrix} .
\end{equation}

Die CRLB ist in diesem Fall also zeitabhängig. Durch Mittelung der erhaltenen Werte über die Sequenzlänge $N$ erhält man dann eine mittlere CRLB.

```python
# compute CRLB for estimating signal x
# CRLB depends on time t

crlbX = np.zeros(len(N))
for cntr, n in enumerate(N):
    tmp = 0.0
    for run in np.arange(1,n+1):
        jacobi = np.array([1, Ta*run, (Ta*run)**2]).reshape(P,1)
        Jextinv = jacobi.T @ Jinv @ jacobi
        tmp += Jextinv[0][0]
    crlbX[cntr] = tmp / n


fig_crlb = plt.figure(figsize=(9,5))

ax1 = fig_crlb.add_subplot(111)
ax1.semilogy(N, crlbX, '-')
plt.xlabel('$N$')
plt.ylabel('CRLB')
plt.title('CRLB für Polynom mit $P=$' + str(P) + ' und AWGN')
plt.grid(True)
plt.legend()
plt.show()
```

## Vergleich mit Least-Squares-Schätzung

Für den betrachteten Fall $\mathbf{y}=\mathbf{H} {\theta} +\mathbf{w}$ lautet die Least-Squares-Schätzung (Methode der kleinsten Fehlerquadrate)

\begin{equation}
    \hat{\mathbf{\theta}} = \left( \mathbf{H}^{\mathrm{T}} \mathbf{H} \right)^{-1} \mathbf{H}^{\mathrm{T}} \cdot \mathbf{y} .
\end{equation}

Den mittleren quadratischen Fehler (MSE: mean square error) der Schätzung erhält man über

\begin{equation}
    \mathrm{MSE}(\theta) 
    = \mathrm{E} \left\{ \left\| \hat{{{\theta}} - {{\theta} \right\|^2 \right\}
    \approx \frac{1}{P} \sum_{p=0}^{P-1} \left\| \hat{\theta}_p - \theta_p \right\|^2 .
\end{equation}

Es sind also mindestens $P$ Schätzungen durchzuführen. Je größer $P$, desto genauer die Schätzung der Schätzvarianz.

Aus den geschätzten Parametern $\hat{\theta}$ lässt sich dann das Signal $x(t_n)$ rekonstruieren. Die Schätzung erhält man über

\begin{equation}
    \hat{x}(t_n) = \sum_{p=0}^{P-1} \hat{\theta}_p t_n^p .
\end{equation}

Auch hier kann der MSE durch

\begin{equation}
    \mathrm{MSE}(X) 
    = \mathrm{E} \left\{ \left\| \hat{\mathbf{x}} - \mathbf{x} \right\|^2 \right\}
    \approx \frac{1}{N} \sum_{n=0}^{N-1} \left\| \hat{x}_n - x_n \right\|^2
\end{equation}

bestimmt werden.


```python
# coefficients of polynomial
P = np.array([1,0.5,-0.1])

# number of runs
Nruns = 1000

# polynomial
Phat = np.zeros((len(N),len(P)))
MSE_P = np.zeros((len(N),Nruns))
t = [None] * len(N)
x = [None] * len(N)
xhat = [None] * len(N)
y = [None] * len(N)
MSE_x = np.zeros((len(N),Nruns))

for cntr, n in enumerate(N):
    t[cntr] = np.arange(n) * Ta
    x[cntr] = P[0] + P[1]*t[cntr] + P[2]*t[cntr]**2

    # system matrix H
    H = np.block([np.ones((n,1)),t[cntr].reshape(n,1),(t[cntr]**2).reshape(n,1)])
    Hpinv = np.linalg.inv(H.transpose() @ H) @ H.transpose()

    xhat[cntr] = np.zeros((n,Nruns))
    for run in np.arange(Nruns):
        awgn = np.random.normal(0,np.sqrt(sigma2_N),n)
        y[cntr] = x[cntr] + awgn
    
        # LS parameter estimation
        Phat[cntr,:] = Hpinv @ y[cntr]
        # estimation error
        MSE_P[cntr,run] = np.sum(np.abs(P-Phat[cntr,:])**2)
    
        # reconstructed x[n]
        xhat[cntr][:,run] = Phat[cntr,0] + Phat[cntr,1]*t[cntr] + Phat[cntr,2]*t[cntr]**2
        # reconstruction error
        MSE_x[cntr,run] = np.mean(np.abs(x[cntr]-xhat[cntr][:,run])**2)

MSE_Pmean = np.mean(MSE_P,axis=1)
MSE_xmean = np.mean(MSE_x,axis=1)
```

```python
fig_example = plt.figure(figsize=(9,5))

ptr = 9
ax2 = fig_example.add_subplot(111)
ax2.plot(t[ptr], xhat[ptr][:,1:100], 'g:', linewidth=0.2)
ax2.plot(t[ptr], x[ptr], 'b-',label='$x[n]$')
ax2.plot(t[ptr], y[ptr], 'r-',label='$y[n]$')
#ax2.plot(np.arange(N[0]) * Ta, xhat[0][:,1], 'g:',label='$\hat{x}[n],N=$+str(N[0])')
#ax2.plot(np.arange(N[2]) * Ta, xhat[2][:,1], 'g--',label='$\hat{x}[n],N=$+str(N[2])')
#ax2.plot(np.arange(N[4]) * Ta, xhat[4][:,1], 'g-',label='$\hat{x}[n],N=$+str(N[4])')
#ax2.plot(np.arange(N[-1]) * Ta, xhat[-1][:,1], 'g.-',label='$\hat{x}[n],N=$+str(N[-1])')
plt.xlabel('$n$')
plt.title('LS-Schätzung $\hat{x}[n]$ für Polynom mit $\mathbf{P}=$' + str(P) + ', $N=$' + str(N[ptr]) + ' und AWGN')
plt.grid(True)
plt.legend()
plt.show()
```

Das obige Bild zeigt das Informationssignal $x[n]$, die verrauschte Beobachtung $y[n]$ und in grün verschiedene LS-Schätzungen $\hat{x}[n]$ für die Sequenzlänge $N=50$. Über die Wahl der Variablen 'ptr' kann die Sequenzlänge im simulierten Bereich variiert werden.

## Vergleich von CRLB und Least-Squares-Schätzung

```python
fig_MSE = plt.figure(figsize=(9,5))

ax3 = fig_MSE.add_subplot(111)
ax3.semilogy(N, MSE_Pmean, 'b-+',label='MSE$(P)$')
ax3.semilogy(N, crlbTotal, 'k-x',label='$CRLB(P)$')
ax3.semilogy(N, MSE_xmean, 'r-s',label='MSE$(x)$')
ax3.semilogy(N, crlbX, 'g-d',label='CRLB$(x)$')
plt.xlabel('$N$')
plt.title('MSE und CRLB für Polynom mit P=' + str(P) + ' und AWGN')
plt.grid(True)
plt.legend()
plt.show()
```

Das obere Bild verdeutlicht, dass die LS-Schätzung des Parametervektors $\mathbf{\theta}$ eine MVU-Schätzung (minimum variance unbiased) darstellt, da sie die CRLB erreicht. Kleinere Abweichungen sind durch den endlichen Stichprobenumfang zu erklären. 

Bei der Schätzung des Signals $x[n]$ liegt hingegen keine MVU-Schätzung vor. Nur asymptotisch für größere $N$ wird die CRLB erreicht.
