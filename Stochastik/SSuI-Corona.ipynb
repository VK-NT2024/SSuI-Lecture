# Bedingte Wahrscheinlichkeiten und Entropien am Beispiel von Corona-Tests

## Corona-Tests und ihre Zuverlässigkeit

Es gibt derzeit drei verschiedene Corona-Tests, welche unterschiedliche Zuverlässigkeiten besitzen. Dabei hängt die absolute Zuverlässigkeit immer auch von der Qualität der Probenentnahme ab. Wird diese nicht fachmännisch durchgeführt, ist das Ergebnis entsprechend ungenau.

* PCR-Tests: (PCR: Polymerase Kettenreaktion) werden im Labor ausgewertet. Im Kern werden durch Erhitzen und Abkühlen sowie die Hinzugabe von ausgewählten Gen-Sequenzen DNA-Stränge aufgelöst und neu gebildet. Die zugeführten Gen-Sequenzen enthalten optische Marker (Fluoreszenz), welche der Test-Auswertung dienen. Bei geringer Zahl von Corona-Viren werden viele Zyklen benögt, um eine gewisse Menge an fluoreszierenden Markern zu erhalten. Bei hoher Anzahl von Viren wird diese Menge schon nach wenigen Zyklen erreicht, so dass die Anzahl der Zyklen ein Maß für die Anzahl nachgewiesener Viren ist. 
* Schnelltests: Schnelltests erlauben eine Auswertung in etwa 15 Minuten, welche daher deutlich schneller aus die Laborauswertung der PCR-Tests ist. Es werden keine Viren sondern Proteine nachgewiesen.
* Nachweis von Antikörpern: Dieser Nachweise wird in der Regel nach erfolgter Infektion durchgeführt und im Folgenden nicht weiter betrachtet.

### Sensitivität 

Bei der Zuverlässigkeit sind Sensitivität und die Spezifität zu unterscheiden. Unter Sensitivität versteht man die Wahrscheinlichkeit, dass ein Test bei einer infizierten Person positiv ausfällt. Mathematisch wird dies über die bedingte Wahrscheinlichkeit 

\begin{equation}
    \Pr \{ \text{Test positiv} \mid \text{Person krank}  \} 
\end{equation}

Die Zuverlässigkeit von PCR-Tests liegt bei der Sensitivität bei über 98%, die von Schnelltests zwischen 95% und 99%. 

### Spezifität

Die Spezifität gibt an, mit welcher Wahrscheinlichkeit eine gesunde Person als gesund erkannt wird, der Test also negativ ausfällt. 

\begin{equation}
    \Pr \{ \text{Test negativ} \mid \text{Person gesund}  \} 
\end{equation}

Die Zuverlässigkeit liegt hier bei PCR-Tests zwischen 93% (niedrige Viren-Konzentration) und 99%. PCR-Tests sind daher häufiger falsch negativ als falsch positiv. Bei Schnelltests liegt die Zuverlässigkeit wieder zwischen 95% und 99%.


## Interpretation der Zuverlässigkeit bei Masseneinzeltests

In der folgenden Analyse werden wir den Zustand einer Person mit dem Prozess $\cal Z$ und der Menge $\mathbb{Z}=\{Z_0=\text{gesund}, Z_1=\text{krank}\}$, das Testergebnis mit dem Prozess $\cal T$ und der Menge $\mathbb{T}=\{T_0=\text{negativ}, T_1=\text{positiv}\}$ beschreiben. Die Sensitivität wird damit zu 

\begin{equation}
    \Pr \{ {\cal T}=T_1 \mid {\cal Z}=Z_1  \} \; ,
\end{equation}

die Spezifität zu

\begin{equation}
    \Pr \{ {\cal T}=T_0 \mid {\cal Z}=Z_0  \} \; .
\end{equation}

Die Interpretation eines Test-Ergebnisses hängt immer stark vom Kontext ab. Über den Satz von Bayes lassen sich Sensitivität und Spezifität folgendermaßen umformen. 

\begin{align}
    \Pr \{ {\cal Z}=Z_1 \mid {\cal T}=T_1 \} 
    &= \Pr \{ {\cal T}=T_1 \mid {\cal Z}=Z_1 \} 
        \cdot \frac{ \Pr \{ {\cal Z}=Z_1 \} }{ \Pr \{ {\cal T}=T_1 \} }
    \\[2ex]
    \Pr \{ {\cal Z}=Z_0 \mid {\cal T}=T_0 \} 
    &= \Pr \{ {\cal T}=T_0 \mid {\cal Z}=Z_0 \}
        \cdot \frac{ \Pr \{ {\cal Z}=Z_0 \} }{ \Pr \{ {\cal T}=T_0 \} }
\end{align}

Die Aussagekraft eines Testergebnisses hängt also immer von den Vortestwahrscheinlichkeiten $\Pr\{ {\cal Z}=Z_1\}=\Pr \{\text{Person krank}\}$ und $\Pr\{ {\cal Z}=Z_0\}=\Pr\{\text{Person gesund}\}$ ab. Bei derzeit (Stand 14.12.2020) knapp 1,4 Mio. bekannten Infizierten in Deutschland können diese Vortestwahrscheinlichkeiten wie unten angegeben angenommen werden. Außerdem werden die Wahrscheinlichkeiten $\Pr\{ {\cal T}=T_1\}=\Pr\{ \text{Test positiv}\}$ und $\Pr\{ {\cal T}=T_0\}=\Pr\{\text{Test negativ}\}$ benötigt. Diese lassen sich über die Beziehungen 

\begin{align}
    \Pr\{ {\cal T}=T_0 \}
    &= \Pr\{ {\cal T}=T_0, {\cal Z}=Z_0 \} + \Pr\{ {\cal T}=T_0, {\cal Z}=Z_1 \} \\
    &= \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_0 \} \cdot \Pr\{ {\cal Z}=Z_0 \}
     + \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_1 \} \cdot \Pr\{ {\cal Z}=Z_1 \}
\end{align}

berechnen.

Mit den oben aufgeführten Zuverlässigkeiten

|    .   |  PCR-Test    | .| Schnell-Test | .|
| ----- | ----- | ----- | ----- | ----- |
| $\Pr\{$ ${\cal T}_{\mathrm{PCR}}$ \| ${\cal Z}\}$ | gesund | krank | gesund | krank |
| positiv |  0.04  |   0.98   | 0.05  |  0.95  |
| negativ |  0.96  |   0.02   | 0.95  |  0.05  |

ergeben sich damit die folgenden Infektionswahrscheinlichkeiten.

|    .  |  PCR-Test | . | Schnell-Test | . |
| ----- | ----- |  ----- |  ----- | ----- |
| $\Pr\{$ ${\cal Z}$ \| ${\cal T}\}$ | positiv | negativ | positiv | negativ |
| gesund |  0.6962  |   0.9996   | 0.7471  |  0.9991  |
| krank |  0.3038  |   0.00037   | 0.2529  |  0.0009  |

Würden mit Massentests alle Einwohner Deutschlands getestet, würden nur ca. 25% bis 30% der positiv getesteten tatsächlich infiziert sein ("true positives"). Daher sind Masseneinzeltests eher kritisch zu sehen. 


```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler

custom_cycler = (cycler(color=['b', 'g', 'r', 'c']))

# Anzahl Einwohner Deutschland
citizens_Germany = 8e7
corona_Germany = 1.4e6

# Vortestwahrscheinlichkeiten [gesund - krank]
Pr_Z_Germany = np.array([1.0 - corona_Germany / citizens_Germany, corona_Germany / citizens_Germany])

print("Die Vortestwahrscheinlichkeiten lauten")
print("Pr{Person krank} = %g und Pr{Person gesund} = %g.\n" %(Pr_Z_Germany[1],Pr_Z_Germany[0]))

# Zuverlässigkeit Pr{T|Z} der Tests (Zeilen: neg,pos x Spalten: gesund,krank)
Pr_T_Z_PCR = np.array([[0.96, 0.02],[0.04, 0.98]])
Pr_T_Z_schnell = np.array([[0.95, 0.05],[0.05, 0.95]])

# Wahrscheinlichkeiten Test-Ergebnisse
Pr_T_PCR_Germany = Pr_T_Z_PCR @ Pr_Z_Germany
Pr_T_schnell_Germany = Pr_T_Z_schnell @ Pr_Z_Germany

print("Der PCR-Test fällt mit einer Wahrscheinlichkeit %g positiv und mit einer Wahrscheinlichkeit %g negativ aus.\n" %(Pr_T_PCR_Germany[1],Pr_T_PCR_Germany[0]))
print("Der Schnelltest fällt mit einer Wahrscheinlichkeit %g positiv und mit einer Wahrscheinlichkeit %g negativ aus.\n" %(Pr_T_schnell_Germany[1],Pr_T_schnell_Germany[0]))
      
# Infektionswahrscheinlichkeiten
Pr_Z_T_PCR_Germany = Pr_T_Z_PCR * np.tile(Pr_Z_Germany,(2,1)) / np.tile(Pr_T_PCR_Germany,(2,1)).transpose()
Pr_Z_T_schnell_Germany = Pr_T_Z_schnell * np.tile(Pr_Z_Germany,(2,1)) / np.tile(Pr_T_schnell_Germany,(2,1)).transpose()


print("Die Infektionswahrscheinlichkeit (Zeilen: neg,pos x Spalten: gesund,krank) beträgt beim PCR-Test")
print(Pr_Z_T_PCR_Germany)
print("und beim Schnelltest")
print(Pr_Z_T_schnell_Germany)
```

## Interpretation anlassbezogener Tests 

In Deutschland wurden bisher vorwiegend Tests beim Vorliegen von Symptomen durchgeführt. Unter der Annahme, dass vor vorhandener Symptomatik in 50% der Fälle eine Corona-Infektion vorliegt ($\Pr\{ {\cal T}=T_1 \mid {\cal Z}=Z_1 \} = 0.5$), ergeben sich die folgenden Infektionswahrscheinlichkeiten.

|   .   | PCR-Test | . | Schnell-Test} | . |
| ----- | ----- |  ----- |  ----- | ----- |
| $\Pr\{$ ${\cal Z}$ \| ${\cal T}\}$ | positiv | negativ | positiv | negativ |
| gesund |  0.0392  |  0.9796  | 0.05  |  0.95  |
| krank  |  0.9608  |  0.0204  | 0.95  |  0.05  |

Bei anlassbezogenen Tests sieht die Infektionswahrscheinlichkeit nun völlig anders aus. Bei positivem Test sind nun 95% - 96% tatsächlich infiziert, bei negativem Testergebnis würde zu zwischen 95% beim Schnelltest und 98% beim PCR-Test keine Infektion vorliegen. 

```python
# symptombezogene Tests [gesund - krank]
Pr_Z_fixed = np.array([0.5,0.5])

print("Die Vortestwahrscheinlichkeiten bei Symptom-bezogenen Tests lauten")
print("Pr{Person krank} = %g und Pr{Person gesund} = %g.\n" %(Pr_Z_fixed[0],Pr_Z_fixed[1]))

# Marginale Wahrscheinlichkeiten Test-Ergebnisse
Pr_T_PCR_fixed = Pr_T_Z_PCR @ Pr_Z_fixed
Pr_T_schnell_fixed = Pr_T_Z_schnell @ Pr_Z_fixed

print("Der Test fällt mit einer Wahrscheinlichkeit %g positiv und mit einer Wahrscheinlichkeit %g negativ aus.\n" %(Pr_T_PCR_fixed[0],Pr_T_PCR_fixed[1]))
      
# Infektionswahrscheinlichkeiten
Pr_Z_T_PCR_fixed = Pr_T_Z_PCR * np.tile(Pr_Z_fixed,(2,1)) / np.transpose(np.tile(Pr_T_PCR_fixed,(2,1)))
Pr_Z_T_schnell_fixed = Pr_T_Z_schnell * np.tile(Pr_Z_fixed,(2,1)) / np.transpose(np.tile(Pr_T_schnell_fixed,(2,1)))


print("Die Infektionswahrscheinlichkeit (Zeilen: neg,pos x Spalten: gesund,krank) beträgt beim anlassbezogenen PCR-Test")
print(Pr_Z_T_PCR_fixed)
print("und beim anlassbezogenen Schnelltest")
print(Pr_Z_T_schnell_fixed)
```

Das obige Beispiel für eine speziell gewählte Infektionswahrscheinlichkeit bei vorhandenen Symptomen soll nun verallgemeinert werden. Dafür wird die Vortestwahrscheinlichkeit (Wahrscheinlichkeit für Corona-Infektion bei vorliegender Symptomatik von Null bis Eins variiert und daraus die jeweiligen Wahrscheinlichkeiten für "false negative" und "true positive" berechnet. Es ergeben sich die unten dargestellten Kurven. Die schwarze senkrechte Linie gibt den Punkt für Massentests in ganz Deutschland an, also für den Fall $\Pr\{{\cal Z}=Z_0\}=0.9825$ und $\Pr\{{\cal Z}=Z_1\}=0.0175$. 

Man erkennt, dass die "true positives" bei sehr geringem Infektionsgeschehen sehr niedrig sind und damit die "false positives" entsprechend hoch. Es werden in diesem Bereich also wie bereits oben beobachtet viele falsche Positiv-Tests generiert. Allerdings steigt die Wahrscheinlichkeit für "true positives" schnell an, wenn die Anzahl der Infizierten steigt. Für sehr hohe Infektionswahrscheinlichkeiten strebt die "true positives"-Rate gegen Eins. In diesem Bereich wird die Aussagekraft negativer Tests kleiner und strebt gegen Null.

Es ist nochmals anzumerken, dass die angegebenen Zuverlässigkeiten unter Laborbedingungen gelten. Bei nicht sachgerechter Durchführung der Tests besteht das Risiko, dass der Abstrich nicht an den eigentlichen Virenzentren vorgenommen wird und damit die Virendichte auf der Probe geringer ist, als es dem Gesundheitszustand entspricht. Demensprechend wird der Anteil nicht erkannter Infektionen unter diesen Bedingungen weiter ansteigen.

```python
# symptombezogene Tests: Wahrscheinlichkeit für "keine Infektion"
Pr_Z0 = np.linspace(0,1,1000)

Pr_Z_T_PCR_variable = np.zeros((2,2,len(Pr_Z0)))
Pr_Z_T_schnell_variable = np.zeros((2,2,len(Pr_Z0)))

for run in np.arange(len(Pr_Z0)):
    Pr_Z_variable = np.array([Pr_Z0[run],1-Pr_Z0[run]])

    # Wahrscheinlichkeiten Test-Ergebnisse
    Pr_T_PCR_variable = Pr_T_Z_PCR @ Pr_Z_variable
    Pr_T_schnell_variable = Pr_T_Z_schnell @ Pr_Z_variable

    # Infektionswahrscheinlichkeiten
    Pr_Z_T_PCR_variable[:,:,run] = Pr_T_Z_PCR * np.tile(Pr_Z_variable,(2,1)) / np.tile(Pr_T_PCR_variable,(2,1)).transpose()
#    Pr_Infektion_PCR_truepositive[run] = tmp[1,1]
#    Pr_Infektion_PCR_truenegative[run] = tmp[0,0]
    Pr_Z_T_schnell_variable[:,:,run] = Pr_T_Z_schnell * np.tile(Pr_Z_variable,(2,1)) / np.tile(Pr_T_schnell_variable,(2,1)).transpose()
#    Pr_Infektion_schnell_truepositive[run] = tmp[1,1]
#    Pr_Infektion_schnell_truenegative[run] = tmp[0,0]

fig1 = plt.figure(figsize=(9,7))
ax1a = fig1.add_subplot(211)
ax1a.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[1,1,:],'b-',label="PCR")
ax1a.plot(1-Pr_Z0,Pr_Z_T_schnell_variable[1,1,:],'r-',label="schnell")
ax1a.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'k-')
#plt.xlabel('$\Pr\{ krank \mid Symptom \}$')
#ax4a.axes.xaxis.set_ticklabels([])
plt.ylabel('true positives')
#plt.xlim((-2,2))
#plt.ylim((-2,2))
plt.grid(True)
plt.legend()

ax1b = fig1.add_subplot(212)
ax1b.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[0,1,:],'b-',label="PCR")
ax1b.plot(1-Pr_Z0,Pr_Z_T_schnell_variable[0,1,:],'r-',label="schnell")
ax1b.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'k-')
plt.xlabel('$\Pr\{ {\cal Z}=Z_1 \}=\Pr\{ krank \mid Symptom \}$')
plt.ylabel('false negatives')
#plt.xticks(np.append(-f_i,f_i))
#plt.xlim((-2,2))
#plt.ylim((-2,2))
plt.grid(True)
plt.legend()
```

## Mehrfachtests

Werden Mehrfachtests je Person durchgeführt, ändert sich die Aussagekraft abermals. In der folgenden Analyse gehen wir davon aus, dass die Tests voneinander unabhängige Ergebnisse entsprechend der oben angegebenen Zuverlässigkeiten $\Pr\{ {\cal T} \mid {\cal Z} \}$ liefern. Dabei gibt es verschiedene Möglichkeiten, die mehrfache Testung auszuwerten. 

* Mehrheitsentscheidung: es wird eine Mehrheitsentscheidung (mindestens 3 Tests erforderlich) gefällt. 
* "1hot-p"-Test: für einen positiven Gesamttest ist es ausreichend, wenn mindestens 1 Test positiv ist.
* "1hot-n"-Test: für einen negativen Gesamttest ist es ausreichend, wenn mindestens 1 Test negativ ist.

### Mehrheitsentscheidung

Bei der Mehrheitsentscheidung liegt ein positives Testergebnis vor, wenn die Mehrzahl der Einzeltests positiv ist. Diese Vorgehensweise ist vergleichbar mit der Decodierung von fehlerkorrigierenden Codes bis zur halben Minimaldistanz. Die Zuverlässigkeiten "true negatives" und "true positives" ergeben sich bei $n$ Tests zu

\begin{align*}
    \Pr \{ {\cal T}_n^{\mathrm{maj}}=T_0 \mid {\cal Z} = Z_0 \} 
    &= \sum_{i=0}^{\left \lfloor \frac{n-1}{2} \right \rfloor} 
        {n \choose i} \cdot \Pr\{ {\cal T}=T_1 \mid {\cal Z}=Z_0 \}^i 
        \cdot \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_0 \}^{(n-i)} \\
    \Pr \{ {\cal T}_n^{\mathrm{maj}}=T_1 \mid {\cal Z} = Z_1 \} 
    &= \sum_{i=0}^{\left \lfloor \frac{n-1}{2} \right \rfloor} 
        {n \choose i} \cdot \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_1 \}^i 
        \cdot \Pr\{ {\cal T}=T_1 \mid {\cal Z}=Z_1 \}^{(n-i)} \; .
\end{align*}

Mit diesen Zuverlässigkeiten können dann wiederum über die Regel von Bayes die Infektionswahrscheinlichkeiten bei gegebenem Testergebnis berechnet werden. Wir erhalten die unten dargestellten Kurven. 

Es ist zu erkennen, dass mit zunehmender Anzahl von Tests pro Person immer genauer indiziert werden kann, ob eine Infektion vorliegt oder nicht. Bei 3 Tests kann die Wahrscheinlichkeit für ein korrektes positives Ergebnis von 30% beim Einzeltest auf 80% angehoben werden. Bei 5 Tests kommt ein falsches positives Ergebnis nur bei sehr geringem Infektionsgeschehen vor. Dabei schneidet der PCR-Test generell besser als der Schnelltest ab.

Ähnliche Aussagen lassen sich über die "false negatives" treffen, also die Ereignisse in denen Infektionen nicht durch die Tests indiziert werden. Eine Besonderheit fällt für $n=4$ Tests auf. Hier verschlechtern sich die Ergebnisse gegenüber $n=3$, es ist also sinnvoll, für eine weitere Steigerung direkt $n=5$ zu wählen.

```python
from scipy.special import binom

# Mehrheitsentscheidung       

# Anzahl Tests
n_Tests = np.array([3,4,5])

Pr_Z_T_PCR_majority = np.zeros((2,2,len(Pr_Z0),len(n_Tests)))
Pr_Z_T_schnell_majority = np.zeros((2,2,len(Pr_Z0),len(n_Tests)))

# Zuverlässigkeit bei Mehrheitsentscheidung
Pr_T_Z_PCR_majority = np.zeros((2,2,len(n_Tests)))
Pr_T_Z_schnell_majority = np.zeros((2,2,len(n_Tests)))

for n in np.arange(len(n_Tests)):
    Pr_tp_PCR_majority = 0.0
    Pr_tn_PCR_majority = 0.0
    Pr_tp_schnell_majority = 0.0
    Pr_tn_schnell_majority = 0.0

    for i in np.arange(0,int((n_Tests[n]-1)/2)+1):
        factor = binom(n_Tests[n],i)
        # true negative
        Pr_tn_PCR_majority += factor * Pr_T_Z_PCR[1,0]**i * Pr_T_Z_PCR[0,0]**(n_Tests[n]-i)
        Pr_tn_schnell_majority += factor * Pr_T_Z_schnell[1,0]**i * Pr_T_Z_schnell[0,0]**(n_Tests[n]-i)
        # true positive
        Pr_tp_PCR_majority += factor * Pr_T_Z_PCR[0,1]**i * Pr_T_Z_PCR[1,1]**(n_Tests[n]-i)
        Pr_tp_schnell_majority += factor * Pr_T_Z_schnell[0,1]**i * Pr_T_Z_schnell[1,1]**(n_Tests[n]-i)

    Pr_T_Z_PCR_majority[:,:,n] = np.array([[Pr_tn_PCR_majority,1-Pr_tp_PCR_majority],[1-Pr_tn_PCR_majority,Pr_tp_PCR_majority]])
    Pr_T_Z_schnell_majority[:,:,n] = np.array([[Pr_tn_schnell_majority,1-Pr_tp_schnell_majority],[1-Pr_tn_schnell_majority,Pr_tp_schnell_majority]])

    
    for run in np.arange(len(Pr_Z0)):
        Pr_Z_majority = np.array([Pr_Z0[run],1-Pr_Z0[run]])

        # Wahrscheinlichkeiten Test-Ergebnisse
        Pr_T_PCR_majority = Pr_T_Z_PCR_majority[:,:,n] @ Pr_Z_majority
        Pr_T_schnell_majority = Pr_T_Z_schnell_majority[:,:,n] @ Pr_Z_majority

        # Infektionswahrscheinlichkeiten
        Pr_Z_T_PCR_majority[:,:,run,n] = Pr_T_Z_PCR_majority[:,:,n] * np.tile(Pr_Z_majority,(2,1)) / np.tile(Pr_T_PCR_majority,(2,1)).transpose()
        Pr_Z_T_schnell_majority[:,:,run,n] = Pr_T_Z_schnell_majority[:,:,n] * np.tile(Pr_Z_majority,(2,1)) / np.tile(Pr_T_schnell_majority,(2,1)).transpose()


fig2 = plt.figure(figsize=(9,7))
ax2a = fig2.add_subplot(211)
ax2a.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[1,1,:],'k-',label="PCR, $n=1$")
ax2a.plot(1-Pr_Z0,Pr_Z_T_schnell_variable[1,1,:],'k--',label="schnell, $n=1$")
ax2a.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[1,1,:,0],'b-',label="PCR, $n=$%g" %(n_Tests[0]))
ax2a.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[1,1,:,1],'g-',label="PCR, $n=$%g" %(n_Tests[1]))
ax2a.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[1,1,:,2],'r-',label="PCR, $n=$%g" %(n_Tests[2]))
ax2a.plot(1-Pr_Z0,Pr_Z_T_schnell_majority[1,1,:,0],'b--',label="schnell, $n=$%g" %(n_Tests[0]))
ax2a.plot(1-Pr_Z0,Pr_Z_T_schnell_majority[1,1,:,2],'r--',label="schnell, $n=$%g" %(n_Tests[2]))
ax2a.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'y-')
plt.ylabel('true positives')
plt.grid(True)
plt.legend()

ax2b = fig2.add_subplot(212)
ax2b.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[0,1,:],'k-',label="PCR, $n=1$")
ax2b.plot(1-Pr_Z0,Pr_Z_T_schnell_variable[0,1,:],'k--',label="schnell, $n=1$")
ax2b.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[0,1,:,0],'b-',label="PCR, $n=$%g" %(n_Tests[0]))
ax2b.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[0,1,:,1],'g-',label="PCR, $n=$%g" %(n_Tests[1]))
ax2b.plot(1-Pr_Z0,Pr_Z_T_PCR_majority[0,1,:,2],'r-',label="PCR, $n=$%g" %(n_Tests[2]))
ax2b.plot(1-Pr_Z0,Pr_Z_T_schnell_majority[0,1,:,0],'b--',label="schnell, $n=$%g" %(n_Tests[0]))
ax2b.plot(1-Pr_Z0,Pr_Z_T_schnell_majority[0,1,:,2],'r--',label="schnell, $n=$%g" %(n_Tests[2]))
ax2b.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'y-')
plt.xlabel('$\Pr\{ krank \mid Symptom \}$')
plt.ylabel('false negatives')
plt.grid(True)
plt.legend()
```

### "1hot-p"- und "1hot-n"-Tests

Bei "1hot-p"-Tests liegt ein positives Testergebnis nur vor, wenn all Tests positiv sind. Hierdurch lässt sich die Fehlerquote positiver Tests extrem verringern. Allerdings geht dies auf Kosten einer höheren "false negatives"-Wahrscheinlichkeit. "true negatives" und "true positives" lauten

\begin{align*}
    \Pr \{ {\cal T}_n^{\mathrm{1hp}}=T_0 \mid {\cal Z} = Z_0 \} 
    &= 1 - \Pr\{ {\cal T}=T_1 \mid {\cal Z}=Z_0 \}^n \\
    \Pr \{ {\cal T}_n^{\mathrm{1hp}}=T_1 \mid {\cal Z} = Z_1 \} 
    &= \Pr\{ {\cal T}=T_1 \mid {\cal Z}=Z_1 \}^n \; .
\end{align*}


Umgekehrt liegt bei "1hot-n"-Tests ein negatives Testergebnis nur vor, wenn alle Tests negativ sind. Hiermit lässt sich die Wahrscheinlichkeit für "false negatives" deutlich verringern, nun auf Kosten einer steigenden "false positives"-Wahrscheinlichkeit. "true negatives" und "true positives" lauten

\begin{align*}
    \Pr \{ {\cal T}_n^{\mathrm{1hn}}=T_0 \mid {\cal Z} = Z_0 \} 
    &= \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_0 \}^n \\
    \Pr \{ {\cal T}_n^{\mathrm{1hn}}=T_1 \mid {\cal Z} = Z_1 \} 
    &= 1 - \Pr\{ {\cal T}=T_0 \mid {\cal Z}=Z_1 \}^n \; .
\end{align*}

Im unten zu findenden Diagramm ist zu erkennen, dass die Verringerung der Fehlerwahrscheinlichkeit bei positiven Tests auf Kosten der Genauigkeit bei negativen Tests geht und umgekehrt.

```python
# "1-hot"-Entscheidungen       

# Anzahl Tests
n_Tests_1hot = np.array([2,3,4,5])

Pr_Z_T_PCR_1hotp = np.zeros((2,2,len(Pr_Z0),len(n_Tests_1hot)))
Pr_Z_T_schnell_1hotp = np.zeros((2,2,len(Pr_Z0),len(n_Tests_1hot)))
Pr_Z_T_PCR_1hotn = np.zeros((2,2,len(Pr_Z0),len(n_Tests_1hot)))
Pr_Z_T_schnell_1hotn = np.zeros((2,2,len(Pr_Z0),len(n_Tests_1hot)))

Pr_T_Z_PCR_1hotp = np.zeros((2,2,len(n_Tests_1hot)))
Pr_T_Z_schnell_1hotp = np.zeros((2,2,len(n_Tests_1hot)))
Pr_T_Z_PCR_1hotn = np.zeros((2,2,len(n_Tests_1hot)))
Pr_T_Z_schnell_1hotn = np.zeros((2,2,len(n_Tests_1hot)))

for n in np.arange(len(n_Tests_1hot)):
    Pr_tp_PCR_1hot = 0.0
    Pr_tn_PCR_1hot = 0.0
    Pr_tp_schnell_1hot = 0.0
    Pr_tn_schnell_1hot = 0.0

    # 1hot-p
    # true negative (at least one test is negative)
    Pr_tn_PCR_1hotp =  1 - Pr_T_Z_PCR[1,0]**n_Tests_1hot[n]
    Pr_tn_schnell_1hotp =  1 - Pr_T_Z_schnell[1,0]**n_Tests_1hot[n]
    # true positive (all tests are positive)
    Pr_tp_PCR_1hotp = Pr_T_Z_PCR[1,1]**n_Tests_1hot[n]
    Pr_tp_schnell_1hotp = Pr_T_Z_schnell[1,1]**n_Tests_1hot[n]

    Pr_T_Z_PCR_1hotp[:,:,n] = np.array([[Pr_tn_PCR_1hotp,1-Pr_tp_PCR_1hotp],[1-Pr_tn_PCR_1hotp,Pr_tp_PCR_1hotp]])
    Pr_T_Z_schnell_1hotp[:,:,n] = np.array([[Pr_tn_schnell_1hotp,1-Pr_tp_schnell_1hotp],[1-Pr_tn_schnell_1hotp,Pr_tp_schnell_1hotp]])

    # 1hot-n
    # true negative (all tests are negative)
    Pr_tn_PCR_1hotn =  Pr_T_Z_PCR[0,0]**n_Tests_1hot[n]
    Pr_tn_schnell_1hotn =  Pr_T_Z_schnell[0,0]**n_Tests_1hot[n]
    # true positive (at least one test is positive)
    Pr_tp_PCR_1hotn = 1 - Pr_T_Z_PCR[0,1]**n_Tests_1hot[n]
    Pr_tp_schnell_1hotn = 1 - Pr_T_Z_schnell[0,1]**n_Tests_1hot[n]

    Pr_T_Z_PCR_1hotn[:,:,n] = np.array([[Pr_tn_PCR_1hotn,1-Pr_tp_PCR_1hotn],[1-Pr_tn_PCR_1hotn,Pr_tp_PCR_1hotn]])
    Pr_T_Z_schnell_1hotn[:,:,n] = np.array([[Pr_tn_schnell_1hotn,1-Pr_tp_schnell_1hotn],[1-Pr_tn_schnell_1hotn,Pr_tp_schnell_1hotn]])


    for run in np.arange(len(Pr_Z0)):
        Pr_Z_1hot = np.array([Pr_Z0[run],1-Pr_Z0[run]])

        # Wahrscheinlichkeiten Test-Ergebnisse bei "1-hot-Entscheidung"
        Pr_T_PCR_1hotp = Pr_T_Z_PCR_1hotp[:,:,n] @ Pr_Z_1hot
        Pr_T_schnell_1hotp = Pr_T_Z_schnell_1hotp[:,:,n] @ Pr_Z_1hot

        Pr_T_PCR_1hotn = Pr_T_Z_PCR_1hotn[:,:,n] @ Pr_Z_1hot
        Pr_T_schnell_1hotn = Pr_T_Z_schnell_1hotn[:,:,n] @ Pr_Z_1hot

        # Infektionswahrscheinlichkeiten
        Pr_Z_T_PCR_1hotp[:,:,run,n] = Pr_T_Z_PCR_1hotp[:,:,n] * np.tile(Pr_Z_1hot,(2,1)) / np.tile(Pr_T_PCR_1hotp,(2,1)).transpose()
        Pr_Z_T_schnell_1hotp[:,:,run,n] = Pr_T_Z_schnell_1hotp[:,:,n] * np.tile(Pr_Z_1hot,(2,1)) / np.tile(Pr_T_schnell_1hotp,(2,1)).transpose()

        Pr_Z_T_PCR_1hotn[:,:,run,n] = Pr_T_Z_PCR_1hotn[:,:,n] * np.tile(Pr_Z_1hot,(2,1)) / np.tile(Pr_T_PCR_1hotn,(2,1)).transpose()
        Pr_Z_T_schnell_1hotn[:,:,run,n] = Pr_T_Z_schnell_1hotn[:,:,n] * np.tile(Pr_Z_1hot,(2,1)) / np.tile(Pr_T_schnell_1hotn,(2,1)).transpose()


fig3 = plt.figure(figsize=(9,7))
ax3a = fig3.add_subplot(211)
ax3a.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[1,1,:],'k-',label="PCR, $n=1$")
#ax3a.plot(1-Pr_Z0,Pr_Z_T_schnell_variable[1,1,:],'k--',label="schnell, $n=1$")
for n in np.arange(len(n_Tests_1hot)):
    ax3a.plot(1-Pr_Z0,Pr_Z_T_PCR_1hotp[1,1,:,n],'-',label="1hot-p, $n=$%g" %(n_Tests_1hot[n]))
    ax3a.plot(1-Pr_Z0,Pr_Z_T_PCR_1hotn[1,1,:,n],'--',label="1hot-n, $n=$%g" %(n_Tests_1hot[n]))
ax3a.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'r-')
ax3a.set_prop_cycle(custom_cycler)
plt.ylabel('true positives')
plt.title('PCR-Tests')
plt.grid(True)
plt.legend()

ax3b = fig3.add_subplot(212)
ax3b.plot(1-Pr_Z0,Pr_Z_T_PCR_variable[0,1],'k-',label="PCR, $n=1$")
for n in np.arange(len(n_Tests_1hot)):
    ax3b.plot(1-Pr_Z0,Pr_Z_T_PCR_1hotp[0,1,:,n],'-',label="1hot-p, $n=$%g" %(n_Tests_1hot[n]))
    ax3b.plot(1-Pr_Z0,Pr_Z_T_PCR_1hotn[0,1,:,n],'--',label="1hot-n, $n=$%g" %(n_Tests_1hot[n]))
ax3b.plot(Pr_Z_Germany[1]*np.ones((2,1)),[0,1],'r-')
ax3b.set_prop_cycle(custom_cycler)
plt.xlabel('$\Pr\{ krank \mid Symptom \}$')
plt.ylabel('false negatives')
plt.grid(True)
plt.legend()
```
